```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#First using pandas we load the CSV file
df = pd.read_csv('supermarket_sales.csv')

#From the original DF we take the Time, and we just took the first 2 values in int 
df['Hour'] = df['Time'].str.slice(0, 2).astype(int)
df['Total'] = df['Total'].round(2)

#We choose data we are looking for make the correlation we made for 4 values crossing 2 by 2 of them, we moved to be a Array in NP
corrTotalHou = df[['Total','Hour']].to_numpy()
corrUnitPRating = df[['Unit price','Rating']].to_numpy()
corrUnitPRating
```

```python
plt.figure(figsize=(8,5))
plt.scatter(corrTotalHou[:,1], corrTotalHou[:,0], alpha=0.6)
plt.xticks(sorted(df['Hour'].unique()))
plt.xlabel('Day Hour')
plt.ylabel('Sales Total')
plt.title('Sales Total/Hour Relation')
```

```python
plt.figure(figsize=(10,8))
plt.scatter(corrUnitPRating[:,1], corrUnitPRating[:,0], alpha=0.6)
plt.xlabel('Unit Price')
plt.ylabel('Rating')
plt.title('Unit Price / Rating Relation')
```
