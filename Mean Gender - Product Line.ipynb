```python
import numpy as np
import pandas as pd

#First using pandas we load the CSV file
df = pd.read_csv('supermarket_sales.csv')

#We have to validate how much Genders does the Dataframe has, this is for validate and see how much of them we have to write
dftotal = df.groupby(['Gender','Product line'])[['Total']].mean().round(2).reset_index()

#After checking the genders the DF has, we just choose the values we will be taking action as saw it the Total
data = df[['Gender','Product line','Total']].to_numpy()

#We select the genders that the Array has choosing the unique ones just the uniques genders are in the Array.
genders = np.unique(data[:,0])

#We select the Product Line that the Array has choosing the unique ones just the uniques Products Line are in the Array.
productLines = np.unique(data[:,1])

#Using nested cicle we iterate in each gender and even in each product Lines that exist, validating and getting the position that is accord the validation
for gender in genders:
    for productL in productLines:
        position = (data[:,0] == gender) & (data[:,1] == productL)

        #Saving this in a new array with the passed validations, getting the mean of the Quantity and Unit Price. Even printing it.
        subset = data[position]
        meanTotal   = np.mean(subset[:,2].astype(float))
        
        print(f"{gender} - {productL}: Total Mean: {meanTotal:.2f}")
```
