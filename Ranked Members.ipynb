```python
import numpy as np
import pandas as pd

#First using pandas we load the CSV file
df = pd.read_csv('supermarket_sales.csv')

#Using pandas we validate the top 5 sales for each member ordering by the quantity
top5Member = df.sort_values(by=['Customer type','Total'], ascending =[True, False]).groupby('Customer type').head(5)

#Obtaning the member and total for each one into Array from the pandas
data = df[['Customer type', 'Total']].to_numpy()

#We checked the unique members that are in the Array, selection the 0 position because member is stored in that position
members = np.unique(data[:, 0])

#Using a cicle for the number of members
for member in members:
    
    #Using a validation we checked the member in the array, thos should be in the main array getting the data wich is in member from all the Array
    position = data[:, 0] == member

    #using a mask we stract the data but related to the member that is iterating the cicle
    subset = data[position]

    #In a new Array we storage all the data from the position 1 that is the total getting all the totals into a new Array
    quantities = subset[:, 1].astype(float)

    #We passed from the  quantities in this case total sorting from Higher to Lowe and just getting the firts 5 because are the top 5 in Total higher
    idTop = np.argsort(quantities)[::-1][:5] 

    #Just select the top 5 from the subset that just contains the member we are iterating.
    topRows = subset[idTop]

    #Print on screen data related to the member
    print(f"\nCustomer Type: {member}")
    print(topRows) 
```
